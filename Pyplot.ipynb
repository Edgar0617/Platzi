```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(-10,10,10)

y = x**2
```

```python
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y,'gx')
plt.show()
```

```python
plt.plot(x,y,'rs-')
plt.show()
```

```python
plt.hist(x)
plt.show()
```

```python
plt.pie(x)
plt.show()
```

```python
plt.scatter(x,y)
plt.show()
```

```python
plt.boxplot(x)
plt.show()
```

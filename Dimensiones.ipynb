```python
import numpy as np
```

Escalar

```python
scalar = np.array(42)
print(scalar)
scalar.ndim
```

Vector

```python
vector = np.array([1,2,3])
print(vector)
vector.ndim
```

Matriz

```python
matriz = np.array([[1,2,3],[4,5,8],[7,1,0],[2,0,1]])
print(matriz)
matriz.ndim
```

Tensor

```python
tensor = np.array([[[1,2,3],[4,5,8],[7,1,0],[2,0,1]],[[1,2,3],[4,5,8],[7,1,0],[2,0,1]]])
print(tensor)
tensor.ndim
```

## Agregar o eliminar dimensiones

```python
vector = np.array([1,2,3],ndmin=10)
print(vector)
vector.ndim
```

```python
hola = np.expand_dims(np.array([1,2,3]),axis=1)
print(hola)
hola.ndim
```

```python
print(vector, vector.ndim)
vector_2 = np.squeeze(vector)
print(vector_2, vector_2.ndim)
```

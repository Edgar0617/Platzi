```python
import pandas as pd
import seaborn as sns
import numpy as np
```

```python
A = np.array([[2, 4], [-1, 2]])
A
```

```python
B = np.array([[1, 2], [0, 3]])
B
```

```python
np.matmul(A, B)
```

## Inversa

```python
Ainversa = np.linalg.inv(A)
Ainversa
```

```python
np.matmul(A, Ainversa)
```

## Valores y Vectores

```python
C = np.array([[1, 2], [1, 0]])
C
```

```python
values, vectors = np.linalg.eig(C)
values, vectors
```

```python
np.matmul(C, vectors.T[1])
```

```python
values[1]*vectors.T[1]
```

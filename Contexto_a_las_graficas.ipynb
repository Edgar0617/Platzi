```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,5,10)
y = np.sin(x)
```

```python
fig, axes = plt.subplots(1,2,figsize=(10,5))
axes[0].plot(x,y,label='$sin(x)$')
axes[0].set_title('Relación x-y')
axes[0].set_xlabel('Eje X')
axes[0].set_ylabel('Eje Y')
axes[0].legend()
axes[1].plot(y,x)
axes[1].set_title('Relación y-x')
axes[1].set_xlabel('Eje Y')
axes[1].set_ylabel('Eje X')
```

```python
plt.plot(x,y,label='sin(x')
plt.title('Este es el titulo')
plt.xlabel('Eje x')
plt.ylabel('Eje y')
plt.legend()
```

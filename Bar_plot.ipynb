```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = np.random.randint(1,50,100)
```

## Histograma

```python
plt.hist(data,bins = 10)
plt.show()
```

```python
plt.hist(data,bins = 10, histtype='step')
plt.style.use('classic')
plt.show()
```

## Boxplot

```python
plt.boxplot(data,vert= False, patch_artist=True)
plt.style.use('classic')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
tips.corr()
```

```python
sns.heatmap(tips.corr())
plt.show()
```

```python
sns.heatmap(tips.corr(),annot=True,cmap='coolwarm')
plt.show()
```

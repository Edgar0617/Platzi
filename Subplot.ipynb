```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(-5,5,15)
y = x**2
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r--')
plt.plot(y,x,'b:')

plt.subplot(1,2,2)
plt.pie(y)
plt.show()
```

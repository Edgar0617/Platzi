```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,5,10)
```

```python
print(plt.style.available)
```

```python
plt.style.use('dark_background')
```

## Matlab

```python
fig, ax = plt.subplots(figsize=(8,8))
ax.plot(x,x+1,'y--')
ax.plot(x,x+2,'b-')
ax.plot(x,x+3,'r')
ax.plot(x,x+4,'go:')
```

## Pyplot

```python
fig, ax = plt.subplots(figsize=(8,8))
ax.plot(x,x+1, color = 'green', alpha = 0.8, linewidth= 5, linestyle = '-', marker = 'o', markersize = 10)
ax.plot(x,x+2, color = 'red',linewidth= 4, linestyle = '--',marker = 'x')
ax.plot(x,x+3, color = 'blue',linewidth= 3)
ax.plot(x,x+4, color = 'orange', linewidth= 2)
```
